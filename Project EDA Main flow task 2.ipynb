```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file into a DataFrame
file_path = "INDIAvi.xlsx"  # Ensure this path matches your file location
data = pd.read_excel(file_path, sheet_name='Sheet1')

# Display the first few rows and basic info
data.head()
data.info()
plt.show()
```

```python
# Remove rows with missing `video_id`
data = data.dropna(subset=['video_id'])

# Fill missing `description` with a placeholder
data['description'].fillna("No description available", inplace=True)

# Verify that there are no remaining missing values
data.isnull().sum()
```

```python
# Create histograms with adjusted bin sizes and log scales
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot views distribution with log scale
sns.histplot(data['views'], bins=40, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Views')
axes[0, 0].set_yscale('log')
axes[0, 0].set_xscale('log')

# Plot likes distribution with log scale
sns.histplot(data['likes'], bins=40, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Likes')
axes[0, 1].set_yscale('log')
axes[0, 1].set_xscale('log')

# Plot dislikes distribution with log scale
sns.histplot(data['dislikes'], bins=40, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Dislikes')
axes[1, 0].set_yscale('log')
axes[1, 0].set_xscale('log')

# Plot comment count distribution with log scale
sns.histplot(data['comment_count'], bins=40, ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Comment Count')
axes[1, 1].set_yscale('log')
axes[1, 1].set_xscale('log')

plt.tight_layout()
plt.show()
```

#1. Exploring Category Distribution:

```python
# Count the number of videos in each category
category_counts = data['category_id'].value_counts()

# Plot the distribution of views across different categories
plt.figure(figsize=(12, 6))
sns.boxplot(x='category_id', y='views', data=data)
plt.title('Views Distribution by Category')
plt.show()
```

```python
#2. Correlation Between Views, Likes, Dislikes, and Comments:
```

```python
# Correlation heatmap
correlation = data[['views', 'likes', 'dislikes', 'comment_count']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Views, Likes, Dislikes, and Comments')
plt.show()
```

#3. Time-based Analysis:

```python
# Convert trending_date to datetime
data['trending_date'] = pd.to_datetime(data['trending_date'], format='%y.%d.%m')

# Group by trending date and sum the views
views_by_date = data.groupby('trending_date')['views'].sum()

plt.figure(figsize=(12, 6))
views_by_date.plot()
plt.title('Total Views Over Time')
plt.xlabel('Trending Date')
plt.ylabel('Total Views')
plt.show()
```

#4. Video Performance Based on Publish Time:

```python
# Convert publish_time to datetime
data['publish_time'] = pd.to_datetime(data['publish_time'])

# Extract hour of publish time
data['publish_hour'] = data['publish_time'].dt.hour

# Group by hour and average views
avg_views_by_hour = data.groupby('publish_hour')['views'].mean()

plt.figure(figsize=(12, 6))
avg_views_by_hour.plot(kind='bar')
plt.title('Average Views by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Views')
plt.show()
```

#5. Top Videos:

```python
# Top 10 videos by views
top_videos = data[['title', 'views', 'likes', 'dislikes', 'comment_count']].sort_values(by='views', ascending=False).head(10)

print(top_videos)
```

#6. Distribution of Engagement Metrics Across Categories:

```python
# Plot likes, dislikes, and comment count by category
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.boxplot(x='category_id', y='likes', data=data, ax=axes[0])
axes[0].set_title('Likes Distribution by Category')

sns.boxplot(x='category_id', y='dislikes', data=data, ax=axes[1])
axes[1].set_title('Dislikes Distribution by Category')

sns.boxplot(x='category_id', y='comment_count', data=data, ax=axes[2])
axes[2].set_title('Comment Count Distribution by Category')

plt.tight_layout()
plt.show()
```

```python
#. Analyzing Relationships Between Variables 
```

```python
# Scatter plots for relationships between views, likes, and comments
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.scatterplot(x='likes', y='views', data=data, ax=axes[0])
axes[0].set_title('Views vs Likes')

sns.scatterplot(x='dislikes', y='views', data=data, ax=axes[1])
axes[1].set_title('Views vs Dislikes')

sns.scatterplot(x='comment_count', y='views', data=data, ax=axes[2])
axes[2].set_title('Views vs Comment Count')

plt.tight_layout()
plt.show()
```
